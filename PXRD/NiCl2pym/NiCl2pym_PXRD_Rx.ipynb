```python
##these just load the required libaries
%matplotlib notebook
import matplotlib.pyplot as plt 
import matplotlib
import numpy as np
from cycler import cycler
```

```python
data = np.loadtxt('NiCl2pym_PXRD_sozizo1_riet_01.dat') #PXRD data file which has been converted to .dat or .xy
rug = np.genfromtxt('NiCl2pym_PXRD_sozizo1_riet_01_rug_marks.dat')
```

```python
fig = plt.figure(100, figsize=(5.3, 2.5))
ax1 = plt.subplot()
ax1.tick_params(axis='both', labelsize=8, direction="in")

Q = 4*np.pi*np.sin(data[:,0]/2/180*np.pi)/1.54

obs = ax1.scatter (Q,data[:,1]+180, s=3, label='$I_{obs}$', color='xkcd:baby poop green')
calc, = ax1.plot (Q,data[:,2]+180, linewidth=1.5, label='$I_{calc}$', color='darkgreen')
diff, = ax1.plot (Q,data[:,3], linewidth=1.5, label='$I_{obs}$-$I_{calc}$', color='lightgrey')
plt.fill_between(Q, (data[:,1] - data[:,4])+180, (data[:,1] + data[:,4])+180, color='limegreen', alpha=0.2)
ax1.plot(rug[:,4], ([150]*len(rug)), '|', ms=7, c='darkgreen')

ax1.set_xlabel('$Q$ (Ã…$^{-1}$)', fontsize=9)
ax1.set_ylabel('$I$', fontsize=9)
ax1.set_xlim((0.76,4.27))
#ax1.set_ylim((-70,760))
plt.yticks([])
plt.xticks(fontsize=8)
ax1.tick_params(axis='both', labelsize=8)
ax1.legend([obs, calc, diff],['$I_{\mathrm{obs\;(ambient)}}$','$I_{\mathrm{calc}}$','$I_{\mathrm{obs}}-I_{\mathrm{calc}}$'], frameon=False, loc='upper right', fontsize=7)
plt.tight_layout()
#plt.savefig('NiCl2pym_PXRD_sozizo1_riet.pdf', bbox_inches='tight', transparent=True) #The output of your plot as an image or a pdf
plt.show()
```

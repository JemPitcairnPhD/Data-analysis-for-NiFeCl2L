```python
%matplotlib notebook
import matplotlib.pyplot as plt 
import matplotlib
import numpy as np
from cycler import cycler
import sys
```

```python
def Q2th (deg,lam): # function to convert 2theta into scattering vector, Q
	
	return 4.0*np.pi*np.sin(deg/360.0*np.pi)/lam
```

```python
data = np.loadtxt('NiCl2btd-d4_tefeyo_PXRD_riet_fit.dat') #PXRD data file which has been converted to .dat or .xy
rug = np.genfromtxt('NiCl2btd-d4_tefeyo_PXRD_riet_fit_rug_marks.dat')
```

```python
fig = plt.figure(100, figsize=(5.3, 2.5))
ax1 = plt.subplot()
ax1.tick_params(axis='both', labelsize=8, direction="in")

Q = 4*np.pi*np.sin(data[:,0]/2/180*np.pi)/1.54

obs = ax1.scatter (Q,data[:,1]+180, s=3, label='$I_{obs}$', color='royalblue')
calc, = ax1.plot (Q,data[:,2]+180, linewidth=1.5, label='$I_{calc}$', color='midnightblue')
diff, = ax1.plot (Q,data[:,3]-900, linewidth=1.5, label='$I_{obs}$-$I_{calc}$', color='lightgrey')
plt.fill_between(Q, (data[:,1] - data[:,4])+180, (data[:,1] + data[:,4])+180, color='cornflowerblue', alpha=0.2)
ax1.plot(rug[:,4], ([-100]*len(rug)), '|', ms=7, c='midnightblue')

ax1.set_xlabel('$Q$ (Ã…$^{-1}$)', fontsize=9)
ax1.set_ylabel('$I$', fontsize=9)
ax1.set_xlim((0.6,4.27))
#ax1.set_ylim((-70,760))
plt.yticks([])
plt.xticks(fontsize=8)
ax1.tick_params(axis='both', labelsize=8)
ax1.legend([obs, calc, diff],['$I_{\mathrm{obs\;(ambient)}}$','$I_{\mathrm{calc}}$','$I_{\mathrm{obs}}-I_{\mathrm{calc}}$'], frameon=False, loc='upper right', fontsize=7)
plt.tight_layout()
#plt.savefig('NiCl2btd-d4_tefeyo_PXRD_riet.pdf', bbox_inches='tight', transparent=True) #The output of your plot as an image or a pdf
plt.show()
```
